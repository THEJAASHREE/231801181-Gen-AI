```python
import nltk
import random
from nltk.corpus import brown
from nltk.tag import hmm
from nltk.probability import LidstoneProbDist

nltk.download('brown')

tagged_sent = list(brown.tagged_sents(categories='news'))

tagged_sent = tagged_sent[:3000]

for sent in tagged_sent:
    sent.append(("UNK", "NN"))

random.shuffle(tagged_sent)

estimator = lambda fd, bins: LidstoneProbDist(fd, 0.1, bins)

trainer = hmm.HiddenMarkovModelTrainer()
hmm_tagger = trainer.train(tagged_sent, estimator=estimator)

sentence = "The system works well".split()

print("Tagged Output:")
print(hmm_tagger.tag(sentence))
```
