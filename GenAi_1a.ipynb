```python
import nltk
import random
from nltk.tag import hmm
from nltk.probability import LidstoneProbDist

train_data = [
    [("I", "PRP"), ("love", "VBP"), ("NLP", "NNP")],
    [("This", "DT"), ("is", "VBZ"), ("easy", "JJ")],
    [("Students", "NNS"), ("learn", "VBP"), ("AI", "NNP")]
]

for sentence in train_data:
    sentence.append(("UNK", "NN"))

random.shuffle(train_data)

estimator = lambda fd, bins: LidstoneProbDist(fd, 0.1, bins)

trainer = hmm.HiddenMarkovModelTrainer()
hmm_tagger = trainer.train(train_data, estimator=estimator)

sentence = "I love AI".split()
tagged_output = hmm_tagger.tag(sentence)

print("Tagged Sentence:")
print(tagged_output)
```
