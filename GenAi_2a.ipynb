```python
!pip install gensim nltk

import nltk
from gensim.models import Word2Vec
from nltk.corpus import brown

nltk.download('brown')
nltk.download('punkt')

sentences = brown.sents()[:5000]

processed = [[w.lower() for w in s if w.isalpha()] for s in sentences if any(w.isalpha() for w in s)]

def train_model(dim, win):
    return Word2Vec(processed, vector_size=dim, window=win, min_count=2, sg=1, epochs=10)

model_50 = train_model(50,5)
model_100 = train_model(100,5)
model_200 = train_model(200,5)

model_w2 = train_model(100,2)
model_w5 = train_model(100,5)
model_w8 = train_model(100,8)

embeddings = {w: model_100.wv[w] for w in model_100.wv.index_to_key}

queries = ['government','money','student','computer']

for word in queries:
    if word in model_100.wv:
        print("\nTop 8 similar words to",word,"(100D):")
        for w,s in model_100.wv.most_similar(word,topn=8):
            print(w,s)

def compare_dim(word):
    n50=[w for w,_ in model_50.wv.most_similar(word,topn=8)]
    n100=[w for w,_ in model_100.wv.most_similar(word,topn=8)]
    n200=[w for w,_ in model_200.wv.most_similar(word,topn=8)]
    print("\nVector Dimension Comparison for",word)
    print("50D:",n50)
    print("100D:",n100)
    print("200D:",n200)

def compare_window(word):
    w2=[w for w,_ in model_w2.wv.most_similar(word,topn=8)]
    w5=[w for w,_ in model_w5.wv.most_similar(word,topn=8)]
    w8=[w for w,_ in model_w8.wv.most_similar(word,topn=8)]
    print("\nWindow Size Comparison for",word)
    print("Window2:",w2)
    print("Window5:",w5)
    print("Window8:",w8)

compare_dim('government')
compare_window('government')

print("\nObservation")
print("Higher vector dimensions capture richer semantic relationships between words.")
print("Smaller window sizes focus on nearby context while larger window sizes capture broader semantic context.")
print("Changing these hyperparameters affects the semantic neighborhood of words in the embedding space.")
```
