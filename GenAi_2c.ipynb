```python
#2c
!pip install gensim

import gensim.downloader as api
import numpy as np
from numpy.linalg import norm

model = api.load("word2vec-google-news-300")

def cosine(a,b):
    return np.dot(a,b)/(norm(a)*norm(b))

def association(w,A,B):
    return np.mean([cosine(model[w],model[a]) for a in A]) - np.mean([cosine(model[w],model[b]) for b in B])

def test_statistic(X,Y,A,B):
    return sum(association(x,A,B) for x in X) - sum(association(y,A,B) for y in Y)

def effect_size(X,Y,A,B):
    num = np.mean([association(x,A,B) for x in X]) - np.mean([association(y,A,B) for y in Y])
    all_words = X + Y
    std = np.std([association(w,A,B) for w in all_words])
    return num/std

career = ["executive","management","professional","corporation","salary","office","business","career"]
family = ["home","parents","children","family","cousins","marriage","wedding","relatives"]
male = ["man","male","boy","brother","he","him","son","father"]
female = ["woman","female","girl","sister","she","her","daughter","mother"]

science = ["science","technology","physics","chemistry","einstein","nasa","experiment","astronomy"]
arts = ["poetry","art","dance","literature","novel","symphony","drama","sculpture"]

stat1 = test_statistic(career,family,male,female)
effect1 = effect_size(career,family,male,female)

stat2 = test_statistic(science,arts,male,female)
effect2 = effect_size(science,arts,male,female)

print("Gender Bias Test (Career vs Family)")
print("Test Statistic:",stat1)
print("Effect Size:",effect1)

print()

print("Gender Bias Test (Science vs Arts)")
print("Test Statistic:",stat2)
print("Effect Size:",effect2)

if effect1>0:
    print("Observation: Career words are more associated with male terms while family words are closer to female terms.")
else:
    print("Observation: No strong gender association detected in career vs family test.")

print()

if effect2>0:
    print("Observation: Science words are more associated with male terms while arts words are closer to female terms.")
else:
    print("Observation: No strong gender association detected in science vs arts test.")
```
