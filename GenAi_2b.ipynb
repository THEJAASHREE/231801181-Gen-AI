```python
#2b
!pip install gensim


import nltk
from gensim.models import Word2Vec
import numpy as np


corpus = [
    "The king rules the kingdom",
    "The queen rules the kingdom",
    "The prince is the son of the king",
    "The princess is the daughter of the queen",
    "The king and queen live in the palace",
    "The prince and princess live in the palace",
    "The palace is in the city",
    "The city has many people",
    "People respect the king",
    "The queen cares for the people",
    "The prince will become king",
    "The princess will become queen"
]

processed_sentences = []

for sentence in corpus:
    words = sentence.lower().split()
    processed_sentences.append(words)

print("Tokenized Sentences:")
print(processed_sentences)


model = Word2Vec(
    sentences=processed_sentences,
    vector_size=50,
    window=3,
    min_count=1,
    sg=1,
    epochs=200
)


vocabulary = model.wv.index_to_key
print("\nVocabulary Generated:")
print(vocabulary)


print("\nWord Embedding Example (king):")
print(model.wv['king'])


query_words = ["king", "queen", "city"]

for word in query_words:
    print("\nTop 5 words similar to:", word)

    similar_words = model.wv.most_similar(word, topn=5)

    for w, score in similar_words:
        print(w, ":", score)


print("\nInterpretation:")
print("Words like king, queen, prince, and princess appear in similar contexts.")
print("Therefore the model places them close in vector space.")
print("This shows that Word2Vec successfully learns semantic relationships.")
```
