```python
import numpy as np

def one_hot_encode(text):
    words = text.lower().split()
    vocab = sorted(set(words))
    word_to_int = {word: i for i, word in enumerate(vocab)}

    matrix = []
    for word in words:
        vector = [0] * len(vocab)
        vector[word_to_int[word]] = 1
        matrix.append(vector)
    return np.array(matrix), vocab

matrix, vocab = one_hot_encode("AI ML DL AI NLP ML")
print("vocab:",vocab)
print(matrix)
```
